Nous voulons déterminer une valeur approchée d'une solution d'équation du type  $f(x)=0$  en utilisant la dichotomie.

a) Définir une fonction `dicho_n(f,a,b,n)` qui prend en entrée une fonction $f$, les extrêmes de l'intervalle $a$ et $b$, et le nombre d'étapes voulues $n$. Elle renvoie en sortie l'encadrement pour la solution approchée par dichotomie, donc les nouvelles valeurs de  $a$  et  $b$ .



```python
def dicho_n(f, a, b, n):
    for i in range(n):
        m = (a + b) / 2
        if f(a) * f(m) > 0:
            a = m
        else:
            b = m
    return (a, b)
```

b) Définir une fonction `dicho_prec(f,a,b,prec)` qui prend en entrée une fonction $f$, les extrêmes de l'intervalle $a$ et $b$, et la précision souhaitée prec. Elle renvoie en sortie l'encadrement pour la solution approchée par dichotomie, donc les nouvelles valeurs de a et b.


```python
def dicho_prec(f, a, b, prec):
    while (b-a)>prec:
        m = (a + b) / 2
        if f(a) * f(m) > 0:
            a = m
        else:
            b = m
    return (N(a),N(b))
```
