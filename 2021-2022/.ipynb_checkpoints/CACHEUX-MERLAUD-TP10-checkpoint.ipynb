# Codage d'une image en couleur

Les images fournies par les appareils photo sont généralement en couleur. Une image en couleur est contituée de trois couches : une couche rouge (R), une couche verte (V), une couche bleue (B). 

Soit $N_y$ le nombre de lignes de l'image et $N_x$ le nombre de colonnes. Le nombre de pixels total est $N=N_x \times N_y$. Chaque couche est une matrice comportant $N_y$ lignes et $N_x$ colonnes. 

Le plus souvent, cette matrice contient des entiers codés sur 8 bits (les valeurs vont de 0 à 255). 
Pour l'image en couleur complète, il y a donc 24 bits par pixels, à multiplier par le nombre de pixels pour obtenir l'occupation totale en mémoire. 

Chaque couche peut être vue comme une image en niveaux de gris. Le niveau 0 est le noir, le niveau 255 est le blanc, le niveau 128 est un gris moyen.

<img src=rgb.JPG></img>

On voit sur la figure le pixel $(5,4)$ dont les niveaux de gris des trois couches sont $(100,20,200)$, ce qui donne une couleur violette. Par convention, on notera en premier l'indice qui repère les colonnes, conformément au repérage des coordonnées d'un point sur un plan $(x,y)$. On remarque néanmoins que l'origine se trouve sur le coin supérieur gauche de l'image.

```python
# rond

luca = imread("luca.png")       

rouge = luca[:,:,0] # matrice de la composante rouge
vert = luca[:,:,1] # matrice de la composante verte
bleu = luca[:,:,2] # matrice de la composante bleu

centrex = luca.shape[0]/2
centrey = luca.shape[1]/2
rayon = luca.shape[0]/2 #ou luca.shape[1]/2 


#on ne peut pas remplir toute la case car les dimensions de l'image ne sont 
#pas les mêmes en x et en y, ce qui implique qu'il faudrait déformer notre cercle

for i in range(luca.shape[0]):#On parcourt toute l'image en abscisses 
    for j in range(luca.shape[1]):#On parcourt toute l'image en ordonnées
        
        distance= math.sqrt((centrex - i)**2 + (centrey - j)**2)
        if (distance > rayon):
            rouge[i,j]=1
            vert[i,j]=1
            bleu[i,j]=1
    
figure(figsize=(5,5))
imshow(rouge,cmap='gray')
```

```python
# tourner 1
# Symétrie par rapport à l'axe des abscisses

from PIL import Image
import pylab as pyl
# ouverture d ’une image au format jpg :
img=Image.open("luca.png")
largeur,hauteur=img.size
img_sym=Image.new("RGB",(largeur,hauteur))

#symétrie axiale selon X
#pour chaque ligne
for y in range(hauteur):
    #pour chaque colonne
    for x in range(largeur):
        p=img.getpixel((x,y))
        # création du pixel correspondant dans la nv image :
        img_sym.putpixel((x,-y+hauteur-1),p)
       
                
img_sym.save("symetrie1.png")
rotation = imread("symetrie1.png")
rouge=rotation[:,:,0]
imshow(rouge,cmap='gray')
        
```

# EXERCICE 3

On veut créer un programme qui réalise un filtre rouge comme pour l'image suivante :

<img src=luca_rouge.png></img>

Il faut travailler sur les trois couches RGB. 



Le programme s'appelera `filtre_rouge`:

    -Il prendra en entréé une image en RGB de votre choix
    -Il affichera l'image
    -Il affichera ses dimensions
    -Il transforme l'image en une matrice tridimensionale
    -Il crée une nouvelle image "vide" avec les mêmes dimensions de celle en entrée
    -Il remplit l'image "vide" avec les mêmes composante de l'image en entrée sauf pour le rouge
    -Il affiche la nouvelle image

```python
#Méthode1
def filtre_rouge1(img_orig):
    image = imread(img_orig)
    show(imshow(image))
    im = np.copy(image) # On fait une copie de l'original
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            r, g, b = im[i, j]
            if (r+220/255 > 1): r=1-220/255 #Si le pixel est au maximum en échelle de rouge
            im[i, j] = (r+220/255,g,b)
    return imshow(im)
```

```python
#METHODE 1
#application à une image de votre choix

image = "luca.png"
filtre_rouge1(image)
```

```python
#METHODE 2

def filtre_rouge(fichier,nom):
    imageorig = imread(fichier)
    show(imshow(imageorig))
    image = Image.open(fichier)
    nouvelle = image.copy()
    for y in range(image.height):
        for x in range(image.width):
            r, g, b = image.getpixel((x,y))
            nouvelle.putpixel((x,y), (r+220,g,b))
    nouvelle.save(nom)
    return nouvelle
```

```python
#METHODE 2
#application à une image de votre choix

fichier = "luca.png"
NouveauNom = "lucarouge.png"

filtre_rouge(fichier,NouveauNom)

filtre = imread("lucarouge.png")
imshow(filtre)
```
