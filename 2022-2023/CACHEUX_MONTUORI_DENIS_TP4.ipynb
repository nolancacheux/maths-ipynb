Afin de justifier le prolongement par continuité en 0,
    
On calcule la limite du taux d’accroissement de la fonction $x$ |$\rightarrow$$ \dfrac{sin x}{x}$ en 0 : 

$$\lim_{x\to 0 } \dfrac{\sin(x)-\sin(0)}{x-0}=\sin'(0) = \cos(0) = 1$$

Sinon, en sachant que la fonction $x$ |$\rightarrow sin x$ est équivalente à la fonction $x$ |$\rightarrow$$ x$ en 0.
Donc $\lim_{x\to 0 }$ $\dfrac{sin x}{x} = \lim_{x\to 0 }\dfrac{x}{x} = 1$ et on retrouve bien le même résultat.

Sachant que a fonction $x$ |$\rightarrow$$ \dfrac{sin x}{x}$ est continue sur $R^*$ , on peut donc poser une fonction f qui est définie sur $R$ : 

$$f(x) = \left\{\begin{array}{cl} \dfrac{\sin(x)}{x} & \text{si } x \neq 0 \\ 1 & \text{si } x = 0 \end{array} \right.$$

ainsi la fonction f est prolongeable par continuité sur l’ensemble des réels.

Afin de justifier le prolongement par continuité en $+\infty$ ,

Nous avons pour $X=nπ$ : $|R(X)| <= \dfrac{2}{n*pi}$ et on doit donner une approximation à $10^-3$ près, on cherche donc une valeur de n tel que : 
$$|R(X)| <= ε$$ $$\ \Longrightarrow \dfrac{2}{n*pi}<= ε$$ $$\Longrightarrow \dfrac{n*pi}{2} >= \dfrac{1}{ε}$$ $$\Longrightarrow \dfrac{n*pi}{2}>= \dfrac{1}{ε}$$ $$\Longrightarrow n>= \dfrac{2}{ε*pi}$$ $$\Longrightarrow n>= \dfrac{2}{ε*pi}$$ $$\Longrightarrow n >=\dfrac{2000}{pi}$$
On choisit donc par exemple n = $\dfrac{2000}{pi}$
On a donc $X = n*pi = 2/ 10^-3 = 2 /ε = 2000$

```python
x = var("x") # symbolic variable 

def F(X):
    return numerical_integral(sin(x)/x, 0, X)[0]

def R(X):
    return numerical_integral(sin(x)/x,X, 1000)[0]


approximation = F(2000) 

reste = R(2000) 


print("Une approximation de I à 10^-3 près est donc :", approximation)
print("La valeur absolue du reste : ", abs(reste),"(qui est bien <= ε )")


import matplotlib.pyplot as plt

X = []

for i in range(90,6300): #
    X.append(abs(R(i)))
    
plt.xlabel("X", size = 16)
plt.ylabel("R(X)", size = 16)

plt.title("A partir du point, |R(X)| <= ε ", 
          fontdict={'family': 'serif', 
                    'color' : 'red',
                    'size': 15})

plt.plot(X) + plt.plot(2000*int(pi),abs(R(2000*int(pi))), marker="o", color="red")
```

La fonction $f(x)$ |$\rightarrow$$\dfrac{1-e^{-x}}{1+x^2}$ est définie sur $R^+$ ainsi la fonction f est prolongeable par continuité sur l’ensemble des réels.

```python
1.2

x = var("x")

f = ( 1-exp(-x) ) / (1+x^2) # function 

def F(X):
    return numerical_integral(( 1-exp(-x) ) / (1+x^2), 0, X)[0]


graph = plot(f, (x,0,150), color="green", legend_label='$f(x)= {1-e^{-x}}/{1+x^2}$') 
graph += plot(F, (x,0,150), color="red", legend_label='$F(X)$')

print("L'intégrale impropre converge vers :",F(1000000))
graph
```

On porte sur le graphe ci-dessus les fonctions $f(x)$ et $F(X)$ et on remarque que lorsque l'on prend une valeur très grande de X, ici nous avons pris X = 150 ,  on constate que l'intégrale impropre semble converger vers 0.95 (avec des X encore plus grand) 

Soit, $R(X) = \displaystyle \int_ {X}^ {+∞} f(x) dx = \displaystyle \int_ {X}^ {+∞} \dfrac{1-e^{-x}}{1+x^2}dx$.

Sur l'intervalle I = 0;+∞ ,$x$ |$\rightarrow$$ \dfrac{1-e^{-x}}{1+x^2}$ et $x$ |$\rightarrow$$ \dfrac{1}{1+x^2}$ sont continues et $\dfrac{1-e^{-x}}{1+x^2} <= \dfrac{1}{1+x^2}$ 

Donc pour A > X, $\displaystyle \int_ {X}^ {A} \dfrac{1-e^{-x}}{1+x^2}dx <= \displaystyle \int_ {X}^ {A}\dfrac{1}{1+x^2}$.
$$$$
Les fonctions étant intégrables sur I, on a bien : $$$$
$$\lim_{A\to +inf } \displaystyle \int_ {X}^ {A} \dfrac{1-e^{-x}}{1+x^2}dx <= \lim_{A\to +inf }  \displaystyle \int_ {X}^ {A}\dfrac{1}{1+x^2}$$
$$$$
$$=> R(X) <= \lim_{A\to +inf }  \displaystyle \int_ {X}^ {A}\dfrac{1}{1+x^2}$$
$$$$
$$=> R(X) <=  \lim_{A\to +inf }  [arctan x]X à A $$
$$$$
$$=> R(X) <=  pi/2 - arctan X$$

Puisque $\lim_{X\to 0 }R(X)=0$ et que $pi/2 - arctan X > 0$ sur l'intervalle des réels, il existe un réel $B$ tel que pour $x > B$ on ait $|R(x)|< pi/2 - arctan X$.

### |R(X)| <=  pi/2 - arctan X

On doit donner une approximation à 10^-3 près, on cherche donc une valeur de X tel que : 
$$$$
$$|R(X)| <= ε => \dfrac{pi}{2} -arctan X <= ε => arctan X >= \dfrac{pi}{2} - ε => X >= tan(\dfrac{pi}{2} - ε) >= 1000 $$
$$$$
Ainsi , pour X >= 1000, |R(X)| <= ε et ainsi on choisit par exemple X = 1000 pour donner notre approximation de I :

```python
x = var("x") # symbolic variable 

def F(X):
    return numerical_integral( ( 1-exp(-x) ) / (1+x^2), 0, X)[0]

def R(X):
    return numerical_integral( ( 1-exp(-x) ) / (1+x^2),X, 1000)[0]


approximation = F(1000)
reste = R(1000) 


print("Une approximation de I à 10^-3 près est donc :", approximation)
print("La valeur absolue du reste : ", abs(reste),"(qui est bien <= ε )")


import matplotlib.pyplot as plt

X = []

for i in range(0,1500): #
    X.append(abs(R(i)))
    
plt.xlabel("X", size = 16)
plt.ylabel("R(X)", size = 16)

plt.title("A partir du point, |R(X)| <= ε ", 
          fontdict={'family': 'serif', 
                    'color' : 'red',
                    'size': 15})

plt.plot(X) + plt.plot(1000,abs(R(1000)), marker="o", color="red")
```
