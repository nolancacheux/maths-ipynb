### Interprétation 

La multiplicité d'un zéro d'une fonction complexe indique combien de fois le zéro est répété dans la fonction. Par exemple, si nous avons la fonction f(z) = (z-1)^2, alors le nombre 1 est un zéro de cette fonction avec une multiplicité de 2, car la fonction peut être écrite comme (z-1) * (z-1). Dans ce cas, le zéro est répété deux fois.

Pour déterminer la multiplicité d'un zéro d'une fonction complexe, il faut d'abord trouver les zéros de la fonction en résolvant l'équation f(z) = 0. Ensuite, il faut examiner la forme générale de la fonction près du zéro en question et compter le nombre de fois que le zéro apparaît dans la décomposition de la fonction.

La visualisation de la fonction peut également être utile pour comprendre son comportement près du zéro en question et déterminer sa multiplicité. Par exemple, si nous avons la fonction f(z) = (z-1)^2, alors la visualisation de cette fonction montre que le zéro 1 est un point d'intersection double, ce qui indique qu'il a une multiplicité de 2.

Les pôles d'une fonction complexe sont les nombres complexes pour lesquels la fonction a une valeur infinie. La multiplicité d'un pôle indique combien de fois le pôle est répété dans la fonction.

La position et la multiplicité des zéros et des pôles peuvent avoir un impact sur les tendances asymptotiques de la fonction. Par exemple, si une fonction a des zéros de multiplicité élevée, cela peut entraîner des comportements oscillants de la fonction autour de ces zéros. Si une fonction a des pôles de multiplicité élevée, cela peut entraîner des comportements exponentiels de la fonction autour de ces pôles.

```python
z = var("z")
h(z) = 1/2 + 3*z/4


for n in [2..200]:
    h += (7/2)*((z^n)/(2^n))
    
complex_plot(h,(-2,2), (-2,2),aspect_ratio=1)
```

### Interprétation 

On détermine donc graphiquement que le domaine de définition de la somme est le disque de rayon de convergence égal à 2, de centre z=0,pour un rayon strictement supérieur à 2, écrire cette somme n'a pas de sens car cela diverge.

Pour z = 2 : on a la série des 2^n/ 2^n qui est la série des 1 qui diverge GROSSIEREMENT donc à la frontière , pour z = 2, écrire cette somme n'a pas de sens !

Le domaine de définition de la somme est le disque ouvert D(0, R) de centre 0 et de rayon R = 2.

On le vérifie avec la règle de d'Alembert (abs(a_n+1/a_n) => R = 1/ (1/2) = 2)

```python
z = var("z")
h(z) = 0

for n in [1..200]:
    h += ((-1)^n) * ((z-3)^n) / (3^(n+1))
    
complex_plot(h,(0,6), (-3,3),aspect_ratio=1)
```

### Interprétation 

On détermine donc graphiquement que le domaine de définition de la somme est le disque de rayon de convergence égal à 3 , de centre z=3 ,pour un rayon strictement supérieur à 3, écrire cette somme n'a pas de sens car cela diverge.

Pour z = 3 : on a la série des ((-1)^n ) / (3^(n+1)) de la forme (-1)^n * $a_n$ avec a_n = 1/ (3^(n+1)) une suite décroissante avec une limite en +infini égale à 0 donc d'après le critère de Leibniz, la série converge et donc la somme existe. 

Le domaine de définition de la somme est le disque fermé D(3, R) de centre 3 et de rayon R = 3.

On le vérifie avec la règle de d'Alembert (abs(a_n+1/a_n) => R = 1/ (1/3) = 3)

```python
z = var("z")
h(z) = 0

for n in [1..100]:
    h += ((-1)^n) * (z^(2*n+1)) / (2*n+1)
    
complex_plot(h,(-1,1), (-1,1),aspect_ratio=1)
```

### Interprétation 

On détermine donc graphiquement que le domaine de définition de la somme est le disque de rayon de convergence égal à 1, de centre z= 0 ,pour un rayon strictement supérieur à 1, écrire cette somme n'a pas de sens car cela diverge.


Le domaine de définition de la somme est le disque fermé D(0, R) de centre 0 et de rayon R = 1.

On le vérifie avec la règle de d'Alembert (abs(a_n+1/a_n) => R = 1/ (1/1) = 1)


Si une fonction est analytique sur un certain domaine, cela signifie qu'elle peut être développée en une série de Taylor à l'ordre infini autour de chaque point du domaine. En d'autres termes, cela signifie que la fonction peut être approchée par une série de Taylor de façon arbitrairement précise pour tous les points du domaine considéré.

Une fonction analytique sur un domaine a certaines propriétés remarquables. Par exemple, elle est continue sur ce domaine et peut être dérivée autant de fois que nécessaire. De plus, une fonction analytique sur un domaine peut être reconstruite à partir de ses valeurs de dérivées en un point quelconque de ce domaine.

Le concept de fonction analytique est important en analyse complexe, car il permet de définir des opérations comme l'intégration et la dérivation pour les fonctions complexes. De plus, il permet de définir des concepts tels que le rayon de convergence d'une série.

Le rayon de convergence d'une série complexe est la distance maximale à partir de l'origine pour laquelle la série converge vers une valeur finie et unique. En d'autres termes, c'est la distance maximale à partir de l'origine pour laquelle la série peut être utilisée pour approcher la valeur de la fonction complexe associée.

Un rayon de convergence peut être déterminé en utilisant les critères de convergence des séries complexes, tels que le critère d'Abel ou le critère de d'Alembert. Si un rayon de convergence est fini, cela signifie que la série converge seulement pour des valeurs de l'argument dans un intervalle autour de l'origine. Si le rayon de convergence est infini, cela signifie que la série converge pour toutes les valeurs de l'argument.

## Interprétation

On peut voir 3 graphs correspondant à <br>
1) la **fonction** $f(z) = sin(pi*Re( z ))*sin(pi*Im(z) )$; <br>
2) la **fonction** $g(z)$ qui ici $=(z^2 + z + 1)/(2 - z)$; <br>
et enfin <br>
3) la **fonction composé** $f°g$, ici $f°g = f(g(z)) = sin(pi*Re((z^2 + z + 1)/(2 - z)))*sin(pi*Im((z^2 + z + 1)/(2 - z)))$. <br> <br>

On peut donc voir que les $zéros$ et $pôles$ de la fonction $g(z)$ sont conservés "après composition".  <br>(ici, les $zéros$ et $pôles$ correspondent aux points quand $|f(z)| = 0$ et $|f(x)| = +\infty$ ) <br><br>
Ici, on peut apercevoir 2 points en zéro (qui sont $(-0.5,0.8)$ et $(-0.5,-0.8)$) et un $pôle$ en $(2,0)$ sur le graph de $g(x)$ (le deuxième) qui se répercutent directement sur le graph de la fonction $f°g$.
<br>
On peut voir graphiquement que le damier original est déformé selon ces variations:<br>
1) Les aires des carrés proches d'un pôles tendent vers 0,<br>
2) les aires des carrés proches d'un $zéros$ sont augmentés.<br>


La transformation conforme est utilisée pour transformer une figure géométrique dans le plan complexe sans changer sa forme géométrique. Cela signifie que les angles et les proportions sont conservés, ce qui permet de simplifier les calculs.

En utilisant cette transformation, on peut ausssi montrer que les points appelés "pôles" et "zéros" sont ici conservés. Un pôle est un point dans le plan complexe où la fonction diverge, tandis qu'un zéro est un point où la fonction s'annule.
