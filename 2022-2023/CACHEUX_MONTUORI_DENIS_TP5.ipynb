On en déduit finalement le nombre N de termes nécessaires pour obtenir une estimation de ζ(α) à ϵ donné : 

```python
S = 0.
sommes_partielles = [ S ]
termes = []
N = 1001

alpha =2

for n in [1..N]:
    S += 1/(n^alpha)
    sommes_partielles.append ( S )
    termes.append(1/(n^alpha))

print("Pour α =2 par exemple :\n")

print("Pour une approximation de la somme avec une erreur d'au plus ϵ, il faut donc choisir (α-1)*N^(α-1) >= 1/ϵ soit : ")

terme = 0
print("Par exemple, on choisit une précision ϵ = 10^-3 (choisir quelque chose d'autre si on veut)  et on reprend alpha =2") 

epsilon = 10^-3

while((alpha-1)*terme^(alpha-1) < 1/epsilon):
    terme+=1
    
print("\nOn a donc N qui doit être supérieur à",terme)


print("\nOn choisit N =",terme+1,"par exemple et on a Sn ≃",S)
print("On voit bien que l'on converge approximativement vers Sn à une erreur de",epsilon,":")

# only plot the first 100 partial sums and terms
img = list_plot ( sommes_partielles [:101])
img += plot (S, (1 ,50) , color ='red',legend_label ="Estimation de S")
img
```

Pour cet exercice, nous devons traiter des valeurs flottante, pas entiere.
Cela explique la présence de S = 0. qui définit S comme un flottant et non un entier  ainsi que le round un peu plus bas

On appliquera la formule $$ \sum_{k=1}^{n}\left [ -(\frac{1}{2b \cdot  (k-1)+1} + ... + \frac{1}{2b\cdot k-1}) + (\frac{1}{2a \cdot  (k-1)} + ... + \frac{1}{2a\cdot k}) \right ] $$

Il y a 3 paramètres : $N, a$ et $b $

$N$ correspond au nombre "d'alternance"  (de bosses et creux) +1  (N=5 correspond à 4 contact avec Pi)

$b$ correspond au nombre de termes impaires à utiliser avant de passer aux termes paires

$a$ est ensuite calculer par $ S = ln\left [ \frac{1}{2}\sqrt{\frac{a}{b}} \right ] $ avec S, notre limite, ici Pi
 donc $a = 4b*e^{2S} $
 
Donc, un cycle vaut $b$ termes impaire + $a$ termes pairs

soit dans notre exercice $10$ termes impairs puis $40e^{2pi}$ termes impaires (environ 21 000)

```python
S = 0.
N = 6
 
#a = 4*b*exp(2*S)      avec S = pi,  b=10 (par rapport à l'information sous le schéma de la consigne)
a = round(40*exp(2*pi))
b = 10

# On initialise notre somme partielle avec une valeur initiale.
sommes_partielles = [S]

# On peut appliquer la formule
for k in [1..N]:
    
    for p in [0..(b-1)]:
        S -= (1/(2*b*k -2*b + 2*p + 1))
        sommes_partielles.append(S)
        
    for l in [1..a]:
        S += (1/(2*a*k -2*a + 2*l))
        sommes_partielles.append(S)

# On affiche "seulement" les 100 000 premiers termes (j'utilise plotjoined=True pour le rendu lisse)
# J'affiche également Pi pour visualisé notre limite de convergence.

img = list_plot(sommes_partielles[:1000000], plotjoined=True, legend_label='Réarrangement de Riemann')
img+= plot(pi, (0 ,len(sommes_partielles)), color='red', linestyle='--', legend_label='Pi limit')
img
```
