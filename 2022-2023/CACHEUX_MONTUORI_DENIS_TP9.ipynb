```python
def all_relations(n):
    ensemble = [(i, j) for i in range(n) for j in range(n)]
    relations = []
    def recursive_relations(ensemble, relation, index):
        if index == len(ensemble):
            relations.append(relation)
            return
        recursive_relations(ensemble, relation, index + 1)
        recursive_relations(ensemble, relation + [ensemble[index]], index + 1)

    recursive_relations(ensemble, [], 0)
    return relations
all_relations(2)
```

Le nombre de relations binaires sur un ensemble E_n x E_n est égal au nombre de façons de choisir une sous-ensemble de E_n x E_n, soit 2^|E_n x E_n|.

Or, |E_n x E_n| = n^2, donc |P(E_n x E_n)| = 2^n^2.

La démonstration peut se faire par récurrence sur n :

Pour n = 1, E_1 = [a] et E_1 x E_1 = [(a, a)]. Le nombre de relations binaires sur E_1 x E_1 sera de 2^1^2 = 2.

Supposons que la déclaration est vraie pour n = k : |P(E_k x E_k)| = 2^k^2.

Pour n = k + 1, E_{k + 1} = [a_1, a_2, ..., a_{k + 1}] et E_{k + 1} x E_{k + 1} = [(a_1, a_1), (a_1, a_2), ..., (a_{k + 1}, a_{k + 1})]. Pour chaque couple d'éléments dans E_{k + 1} x E_{k + 1}, il y aura 2 possibilités : soit inclus, soit exclu. Par conséquent, le nombre total de relations binaires sur E_{k + 1} x E_{k + 1} sera de 2^((k + 1)^2) = 2^k^2 * 2^2 = 2^(k + 1)^2.

Donc, la déclaration est vraie pour tout n, ce qui signifie que |P(E_n x E_n)| = 2^n^2.

Pour vérifier à l'ordinateur, il faut déterminer combien de matrices d'adjacence possibles peuvent être créées pour une relation binaire sur un ensemble E de cardinal n, et ensuite vérifier si chaque matrice est réflexive. Cela peut être fait en utilisant une boucle et en vérifiant la diagonale de chaque matrice pour voir si elle contient uniquement des 1. Cependant, ce serait un processus très long pour des valeurs élevées de n.

Donc nous allons travailler dans E_3 : 

```python
def is_reflexive(A):
    n = A.nrows()
    for i in range(n):
        if A[i, i] == 0:
            return False
    return True

count = 0
for relation in all_relations(3):
    A = mat_adj(3, relation)
    if is_reflexive(A):
        count += 1

print(count)
```

Le nombre de relations réflexives sur un ensemble de cardinalité 3 est donc bien 2^((3 x 3) - 3) = 2^6 = 64 comme nous l'avons démontré.

```python
from itertools import combinations

def is_antisymmetric(A):
    n = A.nrows()
    for i in range(n):
        for j in range(i+1, n):
            if A[i,j] == 1 and A[j,i] == 1:
                return False
    return True


count = 0
for relation in all_relations(3):
    A = mat_adj(3, relation)
    if is_antisymmetric(A):
        count += 1/3
print(count)
```

Il n'est pas garanti que la relation binaire aléatoire Q sur E40 soit réflexive, car elle est générée aléatoirement avec une probabilité de 1/9 pour chaque paire (x, y) de relier x à y. Pour rendre la relation réflexive, nous pouvons ajouter la condition que x soit toujours en relation avec lui-même, c'est-à-dire xRx pour tout x dans E40. La nouvelle relation R est donc réflexive.

De plus pour vérifier que la relation générée est réflexive, nous pouvons utiliser la matrice d'adjacence associée et vérifier si tous les éléments de la diagonale principale sont BIEn eégaux à 1. 
Si ce n'est pas le cas, nous pouvons rendre la relation réflexive en définissant R comme étant l'union de Q et de la relation réflexive. La relation peut donc être rendue réflexive en mettant tous les éléments de la diagonale  à 1 : 

```python
import random

def random_rel(n, p):
    relation = []
    for i in range(n):
        for j in range(i + 1, n):
            if random.random() < p:
                relation.append((i, j))
    return relation

set_random_seed(0)
Q = random_rel(40, 1/9)

#On génère la matrice d'adjacence associée : 
A = mat_adj(40, Q)
#print("\n Matrice d'adjacence de la relation Q :\n")
#print(A)
 

def is_reflexive(A):
    n = A.nrows()
    for i in range(n):
        if A[i, i] == 0:
            return False
    return True

def make_reflexive(A):
    n =A.nrows()
    for i in range(n):
        A[i, i] = 1
    return A

if not is_reflexive(A):
    print("\n\nLa matrice d'adjacence de la relation Q n'est pas réflexive, on la rend ainsi réflexive en mettant tous les éléments de la diagonale à 1 et on obtient ainsi la relation R:\n\n ")
    R = make_reflexive(A)
else:
    print("\n\nLa matrice d'adjacence de la relation Q est réflexive\n\n")
    R = A
    
print(R)
```

Pour vérifier si la relation R est symétrique, il suffit de vérifier si sa matrice d'adjacence est symétrique. De plus, une matrice est symétrique si et seulement si transposée est égale à la matrice elle-même.Si ce n'est pas le cas, on peut la rendre symétrique en changeant les éléments non symétriques pour qu'ils correspondent à leurs équivalents symétriques.

```python
def is_symmetric(A):
    n = A.nrows()
    for i in range(n):
        for j in range(i + 1, n):
            if A[i, j] != A[j, i]:
                return False
    return True

def make_symmetric(A):
    n = A.nrows()
    for i in range(n):
        for j in range(i + 1, n):
            A[j, i] = A[i, j]
    return A

if not is_symmetric(R):
    print("\n\nLa matrice d'adjacence de la relation R n'est pas symétrique, on la rend ainsi symétrique en copiant les éléments symétriques et on obtient ainsi la relation S:\n\n ")
    S = make_symmetric(R)
else:
    print("\n\nLa relation R est déjà symétrique \n\n")
    S = R

print(S)


# Vérifions si S est réflexive
if not is_reflexive(S):
    print("\n\nLa matrice d'adjacence de la relation S n'est plus réflexive")
else:
    print("\n\nLa matrice d'adjacence de la relation S est toujours réflexive")
```
